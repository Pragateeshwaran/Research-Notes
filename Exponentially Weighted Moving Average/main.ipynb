```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
datapoints_x = np.random.normal(loc=0, scale=1, size=100)    
datapoints_y = np.random.normal(loc=0, scale=1, size=100)   
```

```python
sorted_indices = np.argsort(datapoints_x)
datapoints_x_sorted = datapoints_x[sorted_indices]
datapoints_y_sorted = datapoints_y[sorted_indices]
```

```python
plt.scatter(datapoints_x, datapoints_y, alpha=0.7, edgecolors='k')
plt.grid(True)
plt.show()
```

```python
B = 0.9
EWMA = np.zeros(len(datapoints_x))
for i, (x, y) in enumerate(zip(datapoints_x_sorted, datapoints_y_sorted)):
    if i == 0:
        EWMA[i] = y   
    else:
        EWMA[i] = B * EWMA[i-1] + (1 - B) * y 
plt.scatter(datapoints_x_sorted, datapoints_y_sorted, alpha=0.7, edgecolors='k')
plt.plot(datapoints_x_sorted, EWMA, color='r')
plt.grid(True)
plt.show()
```
